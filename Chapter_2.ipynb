```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
%matplotlib inline
```

### Q: 2.4

```python
NUM_SAMPLES = 10000
a = 0
b = 1

samples = np.random.uniform(a, b, NUM_SAMPLES)
expectations = []

for i in range(1, NUM_SAMPLES + 1):
    sub_sample = samples[0:i] 
    expectations.append(sum([np.sin(s) for s in sub_sample]) / len(sub_sample))

plt.ylim(min(expectations) - 0.01, max(expectations) + 0.01)
plt.xscale('log')
plt.xlabel('Number of Samples')
plt.ylabel('Expectation')
plt.title('Approximation of the expectation of $sin(y)$ under $U(0, 1)$')
plt.plot(np.arange(1, NUM_SAMPLES + 1), expectations)
plt.axhline(y=0.4597, color='red', linestyle='--')
plt.show()
```
